```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
#1
df['Pclass'].value_counts().plot(kind='bar')
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Passenger Count by Class")
plt.show()
```

```python
#2
df['Age'].hist(bins=20)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
#3
df['Survived'].value_counts().plot(kind='pie', autopct="%1.1f%%")
plt.title("Survival Proportion")
plt.ylabel("")     # remove y-label
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
#4
# Convert species names to numerical categories for coloring
species_numeric = pd.factorize(df['Species'])[0]

plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], c=species_numeric)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Width (Colored by Species)")
plt.show()
```

```python
#5
df['PetalLengthCm'].hist(bins=20)
plt.xlabel("Petal Length")
plt.ylabel("Count")
plt.title("Petal Length Distribution")
plt.show()
```

```python
#6
avg_pw = df.groupby('Species')['PetalWidthCm'].mean()

avg_pw.plot(kind='bar')
plt.xlabel("Species")
plt.ylabel("Average Petal Width")
plt.title("Avg Petal Width by Species")
plt.show()
```

```python
#10
df = pd.read_csv("winequality-red.csv")

df.head()
```
